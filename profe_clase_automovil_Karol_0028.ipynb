#### Clase automovil con 3 atributos funciones leerdatos() mostrardatos()

```python
class automovil:
    def __init__(self,marca,placa,precio):
        self.marca=marca
        self.placa=placa
        self.precio=precio

    def leerdatos(self):
        self.marca=input("ingrese la marca del automovil: ")
        self.placa=input("ingrese la placa del automovil: ")
        self.precio=input("ingrese el precio del automovil: ")
        return self.marca,self.placa,self.precio

    def mostrardatos(self):
        print("marca: ",self.marca)
        print("placa: ",self.placa)
        print("precio: ",self.precio)
        return self.marca,self.placa,self.precio
#fin de clase automovil


print("carro numero 1")
carro1 = automovil("", "", "") # creando objeto carro 1 con valores iniciales
carro1.leerdatos()
carro1.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*")
print("carro numero 2")
carro2 = automovil("", "", "") # creando objeto carro 2 con valores iniciales
carro2.leerdatos()
carro2.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*")
print("carro numero 3")
carro3 = automovil("", "", "") # creando objeto carro 3 con valores iniciales
carro3.leerdatos()
carro3.mostrardatos()
print("sofware desarrollado por Karol Castillo NL = 9")
```
